## Data Wrangling 

### Load data

```python
# Read data from file
raw = pd.read_csv("PISG_clean.csv")

# Inspect first 3 rows
raw.head(3)
```

### Reshape data

```python
# Guesses: How many eggs may have golden tickets?
guess = raw[["participant", "condition"] + ["guess_" + str(i) for i in range(1, 13)]]

# Decision: Will you buy a new egg (1) or skip it (0)?
decision = raw[["participant"] + ["buy_" + str(i) for i in range(1, 13)]]

# Reshape guesses
guess_long = pd.melt(
    guess,
    id_vars=["participant", "condition"],
    value_vars=["guess_" + str(i) for i in range(1, 13)],
    var_name="trial",
    value_name="guess",
)

# Reshape decisions
decision_long = pd.melt(
    decision,
    id_vars=["participant"],
    value_vars=["buy_" + str(i) for i in range(1, 13)],
    var_name="trial",
    value_name="buy",
)

# Drop columns that are also in guess_long
decision_long.drop(["trial", "participant"], axis=1, inplace=True)

# Merge guess and decision data
merged = pd.concat([guess_long, decision_long], axis=1)

# Replace trial names with the actual design
merged["trial"].replace(
    {
        "guess_1": "6_6_1",
        "guess_2": "6_6_0",
        "guess_3": "6_5_1",
        "guess_4": "6_5_0",
        "guess_5": "6_2_2",
        "guess_6": "6_1_1",
        "guess_7": "6_6_6",
        "guess_8": "6_6_5",
        "guess_9": "6_5_5",
        "guess_10": "6_5_4",
        "guess_11": "6_2_0",
        "guess_12": "6_1_0",
    },
    inplace=True,
)

# Split trial names into 3 columns: total # of eggs laid, checked, and had a ticket
merged[["laid", "checked", "hit"]] = merged["trial"].str.split("_", expand=True)

# Convert strings to numerics
merged["laid"] = pd.to_numeric(merged["laid"])
merged["checked"] = pd.to_numeric(merged["checked"])
merged["hit"] = pd.to_numeric(merged["hit"])
```

```python
# Put each condition in a separate dataframe
check = merged.loc[merged["condition"] == "check"]
both = merged.loc[merged["condition"] == "both"]
```

```python
# Calculate check and hit rates
both["check_rate"] = both["checked"] * 1.0 / both["laid"]
both["hit_rate"] = both["hit"] * 1.0 / both["checked"]
both["hit_inferred"] = both["guess"] * 1.0 / 6
```

```python
# In the check condition, no hit rate was available
check.drop(["hit"], axis=1, inplace=True)


check["trial"].replace(
    {
        "6_6_1": "6_6",
        "6_6_0": "6_6",
        "6_6_6": "6_6",
        "6_6_5": "6_6",
        "6_5_1": "6_5",
        "6_5_0": "6_5",
        "6_5_5": "6_5",
        "6_5_4": "6_5",
        "6_2_2": "6_2",
        "6_2_0": "6_2",
        "6_1_1": "6_1",
        "6_1_0": "6_1",
    },
    inplace=True,
)

# Split trial names into 2 columns: total # of eggs laid and checked
check[["laid", "checked"]] = check["trial"].str.split("_", expand=True)

# Convert strings to numerics
check["laid"] = pd.to_numeric(check["laid"])
check["checked"] = pd.to_numeric(check["checked"])
```

```python
# Add check rate
check["check_rate"] = check["checked"] * 1.0 / check["laid"]

# And inferred hit rate
check["hit_inferred"] = check["guess"] * 1.0 / 6
```
